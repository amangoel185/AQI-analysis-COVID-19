```
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('AQI-analysis-COVID-19-data.csv')
df['local'] = pd.to_datetime(df['local'])
df.head()

fig = plt.figure(figsize=(11, 7))

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.set_title('Trend of CO Values in the North Campus over the Past 3 Months')
ax1.set_xlabel('Date')
ax1.set_ylabel('Value of CO (in µg/m³)')
ax1.plot(df['local'], df['co'], color='green')
```

```
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
%matplotlib inline 
   
df = pd.read_csv('AQI-analysis-COVID-19-data.csv')
df['Day'] = pd.to_datetime(df['local'])
df['Day'] = df['Day'].dt.date

df.head()  

df.boxplot(by ='Day', column =['co'], grid = False, figsize=(27, 18),rot=90, fontsize = 'large')
```
